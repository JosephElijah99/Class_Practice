```python
# QUESTION 11: GROUP CARS BY MANUFACTURERS AND ANALYZE THEEIR AVERAGE PRICE AND FUEL EFFICIENCY

grouped_data = df.groupby('Manufacturer').agg({
    'Price_in_thousands': 'mean',
    'Fuel_efficiency': 'mean'
}).reset_index()
grouped_data
```

```python
# VISUALIZING RELATIONSHIP BETWEEN PRICE_IN_THOUSANDS AND MANUFACTURERS
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Manufacturer', y='Price_in_thousands', color='skyblue')
plt.title('Price_in_thousands vs. Manufacturers')
plt.xlabel('Manufacturers')
plt.ylabel('Price_in_thousands')
plt.xticks(rotation=90)
plt.show()
```

```python
# VISUALIZING RELATIONSHIP BETWEEN FUEL EFFICIENCY AND MANUFACTURERS
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Manufacturer', y='Fuel_efficiency', color='lightgreen')
plt.title('Average Fuel Efficiency by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Fuel Efficiency')
plt.xticks(rotation=90)
plt.show()
```

```python
# QUESTION 13: CAR MODEL WITH HIGHEST RESALE VALUE COMPARED TO INITIAL PRICE

df['Percentage_Change'] = ((df['year_resale_value'] - df['Price_in_thousands']) / df['Price_in_thousands']) * 100
highest_resale_model = df.loc[df['Percentage_Change'].idxmax()]

print("Car model with the highest resale value compared to its initial price:")
print(highest_resale_model)
```

Car Model with the Highest Resale Value is the 'FOCUS' made by FORD MANUFACTURERS

```python
# QUESTION 14: TOP 3 FUEL EFFICIENT CARS WITH ENGINE SIZE ABOVE 3.5 LITRES

# Filtering cars with engine sizes above 2.5 litres
filtered_df = df[df['Engine_size'] > 2.5]
filtered_df
```

```python
# Sorting by fuel efficiency and selecting the top 3
top_3_fuel_efficient_cars = filtered_df.sort_values(by='Fuel_efficiency', ascending=False).head(3)

print("Top 3 fuel-efficient cars with engine size above 2.5 litres:")
print(top_3_fuel_efficient_cars['Model'])
```

```python
# QUESTION 15: WHICH LEXUS MODEL HAS THE HIGHEST HORSEPOWER

# Filtering for Lexus models
lexus_df = df[df['Manufacturer'] == 'Lexus']
lexus_df
```

```python
# Finding the Lexus model with the highest horsepower
highest_hp_lexus_model = lexus_df.loc[lexus_df['Horsepower'].idxmax()]

print("Lexus model with the highest horsepower:")
print(highest_hp_lexus_model['Model'])
```
